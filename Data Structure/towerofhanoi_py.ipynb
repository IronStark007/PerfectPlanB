# Program for mathamatical puzzle - TOWER OF HANOI

```
def remove(lst, p):
    return lst.pop(p)


def add(lst, p):
    return lst.insert(0, p)


def sep(lst):
    snd = []
    for b in lst:
        sp = b.split()
        snd.append(int(sp[1]))
    return snd


def disk(lst):
    e = lst.copy()
    lst.clear()
    for a in e:
        lst.append(f"Disk {a}")
    return lst


def conint(lst):
    com = []
    for b in lst:
        sp = b.split()
        com.append(int(sp[1]))
    return com


def covstr(st):
    lst = [st]
    return lst


def compare(lst1, lst2):
    if lst2 == []:
        ap = remove(lst1, 0)
        conint(lst2)
        add(lst2, ap)
        con = covstr(ap)
        conint(con)
        return True
    else:
        for a in lst1:
            for b in lst2:
                if a > b:
                    return False
                else:
                    ap = remove(lst1, 0)
                    conint(lst2)
                    add(lst2, ap)
                    con = covstr(ap)
                    conint(con)
                    return True


if __name__ == '__main__':

    print("Welcome to the Tower of Hanoi Game")
    n = input("How many disk you want to play with :\n")
    al = []
    bl = []
    cl = []
    move = 0
    try:
        num = int(n)
    except:
        num = -1
        print("Sorry, you have to type only numbers 3,4,5 etc")
    if num >= 3:
        for i in range(num):
            i += 1
            al.append(i)
        dic = {"A": disk(al), "B": bl, "C": cl}
        print(f"OK!,There are three rods namely:A,B,C and {n} disks in the rod A")
        print(dic)
        print("There are total 3 steps for a successful move")
        print("1)Select the disk\n2)Select the disk which you want to move\n"
              "3)Enter the rod where you want it move to")
        print("Lets start the game")
        while len(cl) <= 2:
            move += 1
            r1 = input("Select the rod :")
            d = input("Select the disk :")
            try:
                dsk = int(d)
            except:
                dsk = -1
                print("Sorry,enter only 1, 2 or 3 for disk selection")
            r2 = input("Select where to move :")
            if dsk == 1:
                if r1.lower() == "a" and r2.lower() == "b":
                    if compare(al, bl):
                        dic.update({"B": bl})
                        dic.update({"A": al})
                        print("Successfully moved")
                        print("Updated disk :", dic)
                    else:
                        print("Sorry,can move larger disk on top of smaller disk")
                        print("Disk :", dic)
                elif r1.lower() == "a" and r2.lower() == "c":
                    if compare(al, cl):
                        dic.update({"C": cl})
                        dic.update({"A": al})
                        print("Successfully moved")
                        print("Updated disk :", dic)
                    else:
                        print("Sorry,can move larger disk on top of smaller disk")
                        print("Disk :", dic)
                elif r1.lower() == "b" and r2.lower() == "a":
                    if compare(bl, al):
                        dic.update({"A": al})
                        dic.update({"B": bl})
                        print("Successfully moved")
                        print("Updated disk :", dic)
                    else:
                        print("Sorry,can move larger disk on top of smaller disk")
                        print("Disk :", dic)
                elif r1.lower() == "b" and r2.lower() == "c":
                    if compare(bl, cl):
                        dic.update({"B": bl})
                        dic.update({"C": cl})
                        print("Successfully moved")
                        print("Updated disk :", dic)
                    else:
                        print("Sorry,can move larger disk on top of smaller disk")
                        print("Disk :", dic)
                elif r1.lower() == "c" and r2.lower() == "a":
                    if compare(cl, al):
                        dic.update({"C": cl})
                        dic.update({"A": al})
                        print("Successfully moved")
                        print("Updated disk :", dic)
                    else:
                        print("Sorry,can move larger disk on top of smaller disk")
                        print("Disk :", dic)
                elif r1.lower() == "c" and r2.lower() == "b":
                    if compare(cl, bl):
                        dic.update({"B": bl})
                        dic.update({"C": cl})
                        print("Successfully moved")
                        print("Updated disk :", dic)
                    else:
                        print("Sorry,can move larger disk on top of smaller disk")
                        print("Disk :", dic)
                elif r1.lower() == "a" and r2.lower() == "a":
                    print("Sorry,cannot select and move to the same rod,Try again")
                elif r1.lower() == "b" and r2.lower() == "b":
                    print("Sorry,cannot select and move to the same rod,Try again")
                elif r1.lower() == "c" and r2.lower() == "c":
                    print("Sorry,cannot select and move to the same rod,Try again")
                else:
                    print("Sorry you have to select only from A, B and C")
            elif num >= dsk > 1:
                print("Sorry cant move the lower disk you have to choose the upper one")
        print(f"GAME OVER!!!\nCongrats you have completed the game in {move} steps")
    else:
        print("You have to select disk greater or equal to 3 to start the game")
```
