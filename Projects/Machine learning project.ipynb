```python
# importing pandas and numpy
import pandas as pd 
import numpy as np 

# importing linear regression from sklearn
from sklearn.linear_model import LinearRegression

#importing matplotlib
import matplotlib.pyplot as plt
```

```python
# loading the dataset using pandas
df=pd.read_excel('lrdata.xlsx')

# printing first five rows of data
df.head()
```

```python
# initialising x and y variables
X=df[['cases','increasingrate','decreasingrate']]
Y=df['time']
```

```python
# creating to linear reg func to train the model
def linearreg(x,y):
    reg=LinearRegression()
    reg.fit(x,y)
    return reg
```

```python
# importing train-test split to splitting the data into train and test
from sklearn.model_selection import train_test_split

# splitting the data
x_train,x_test,y_train,y_test=train_test_split(X,Y)
```

```python
# training the train data using func defined 
reg= linearreg(x_train,y_train)
```

```python
# Calculating the accuracy on test data
y_pred_train=reg.predict(x_test)
acc=reg.score(x_test,y_test)
print('Accuracy:',round(acc*100,2))
```

```python
# plotting the graph of data
plt.plot(X,reg.predict(X),'.')
plt.legend(labels=['cases','increasingrate','decreasingrate'])
plt.xlabel('cases rate')
plt.ylabel('time')
plt.show();
```
