```python
%matplotlib inline
# importing the mplot3d toolkit from the mpl_toolkits
from mpl_toolkits import mplot3d

#importing numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating the 3d axis graph
fig = plt.figure()
ax = plt.axes(projection='3d')  #for 3d graph

# creating lines for x, y and z-axis
z_axis = np.linspace(0, 15, 1000)
x_axis = np.sin(z_axis)
y_axis = np.cos(z_axis)

# putting the line in the graph
ax.plot3D(x_axis, y_axis, z_axis, 'black')

# creating scattered plot data for the line
z_scatter = 15 * np.random.random(100)
x_scatter = np.sin(z_scatter) + 0.1 * np.random.randn(100)
y_scatter = np.cos(z_scatter) + 0.1 * np.random.randn(100)

# putting the scattered data in the graph
ax.scatter3D(x_scatter, y_scatter, z_scatter, c=z_scatter, cmap='Oranges')
```

```python
# creating #3D contour plot using mpltoolkit


# defining x, y and z data
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)  #for coordinate matrices
Z = f(X, Y)

# creating 3D contour plot
fig = plt.figure()
ax = plt.axes(projection='3d')  #for 3D graph
ax.contour3D(X, Y, Z, 50, cmap='gray')

# setting the labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
#Creating Surface triangulation graph

# creating x y z and r data
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))  #for contiguous flattened array.
y = np.ravel(r * np.cos(theta))
z = f(x, y)  #function we created earlier

#creating figure
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(x, y, z, c=z, cmap='RdBu', linewidth=0.5)
```
